```python
import pandas as pd
from collections import Counter
from konlpy.tag import Okt
```

```python
# CSV 파일 읽기
df = pd.read_csv('누메로도스.csv')

# 별점이 4점 이상인 리뷰 필터링
filtered_reviews = df[df['rating'] >= 4]['review'].dropna().tolist()

# 형태소 분석기 초기화
okt = Okt()

# 모든 필터링된 리뷰에서 명사 추출
nouns = []
for review in filtered_reviews:
    nouns.extend(okt.nouns(review))

# 단어 빈도수 세기
word_counts = Counter(nouns)

# 상위 10개 단어 추출
top_10_words = word_counts.most_common(10)

# 데이터 프레임으로 변환
top_10_df = pd.DataFrame(top_10_words, columns=['word', 'count'])

# 결과 출력
print(top_10_df)

# CSV 파일로 저장
top_10_df.to_csv('누메로도스_키워드_4점이상.csv', index=False, encoding='utf-8-sig')

```

```python
# 시각화
plt.figure(figsize=(10, 6))
plt.bar(top_10_df['word'], top_10_df['count'], color='skyblue')
plt.title('Top 10 Korean Keywords (Rating >= 4)')
plt.xlabel('Words')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```
